```python
PATIENT_ID = '00006'
TRAIN_DATASET_PATH = '/data/segmentation/train'
#VALIDATION_DATASET_PATH = 'BraTS2020_ValidationData/MICCAI_BraTS2020_ValidationData'
```

```python
test_image_flair = nib.load(os.path.join(TRAIN_DATASET_PATH, PATIENT_ID, PATIENT_ID + '_flair.nii.gz')).get_fdata()
print(test_image_flair.min(), test_image_flair.max())
```

```python
#Scalers are applied to 1D so let us reshape and then reshape back to original shape. 
test_image_flair=scaler.fit_transform(test_image_flair.reshape(-1, test_image_flair.shape[-1])).reshape(test_image_flair.shape)
print(test_image_flair.min(), test_image_flair.max())
```

```python
test_image_flair.shape
```

```python
test_image_t1=nib.load(os.path.join(TRAIN_DATASET_PATH, PATIENT_ID, PATIENT_ID + '_t1.nii.gz')).get_fdata()
test_image_t1=scaler.fit_transform(test_image_t1.reshape(-1, test_image_t1.shape[-1])).reshape(test_image_t1.shape)

test_image_t2=nib.load(os.path.join(TRAIN_DATASET_PATH, PATIENT_ID, PATIENT_ID + '_t2.nii.gz')).get_fdata()
test_image_t2=scaler.fit_transform(test_image_t2.reshape(-1, test_image_t2.shape[-1])).reshape(test_image_t2.shape)

test_image_t1ce=nib.load(os.path.join(TRAIN_DATASET_PATH, PATIENT_ID, PATIENT_ID + '_t1ce.nii.gz')).get_fdata()
test_image_t1ce=scaler.fit_transform(test_image_t1ce.reshape(-1, test_image_t1ce.shape[-1])).reshape(test_image_t1ce.shape)

test_mask=nib.load(os.path.join(TRAIN_DATASET_PATH, PATIENT_ID, PATIENT_ID + '_seg.nii.gz')).get_fdata()
test_mask=test_mask.astype(np.uint8)
```

```python
print(np.unique(test_mask))  #0, 1, 2, 4 (Need to reencode to 0, 1, 2, 3)
test_mask[test_mask==4] = 3  #Reassign mask values 4 to 3
print(np.unique(test_mask)) 
```

```python
n_slice=random.randint(0, test_mask.shape[2])
n_slice = 64
print("slice nr.", n_slice)

plt.figure(figsize=(12, 8))

plt.subplot(151)
plt.imshow(test_image_t1[:,:,n_slice], cmap='gray')
plt.title('Image t1')

plt.subplot(152)
plt.imshow(test_image_t2[:,:,n_slice], cmap='gray')
plt.title('Image t2')

plt.subplot(153)
plt.imshow(test_image_t1ce[:,:,n_slice], cmap='gray')
plt.title('Image t1ce')

plt.subplot(154)
plt.imshow(test_image_flair[:,:,n_slice], cmap='gray')
plt.title('Image flair')

plt.subplot(155)
plt.imshow(test_mask[:,:,n_slice])
plt.title('Mask')
plt.show()
```

```python
##################################################
#PART 2: Explore the process of combining images to channels and divide them to patches
#Includes...
#Combining all 4 images to 4 channels of a numpy array.
#
################################################
#Flair, T1CE, annd T2 have the most information
#Combine t1ce, t2, and flair into single multichannel image
```

```python
combined_x = np.stack([test_image_t1, test_image_t2, test_image_t1ce, test_image_flair], axis=3)
combined_x.shape
```

```python
#Crop to a size to be divisible by 64 so we can later extract 64x64x64 patches. 
#cropping x, y, and z
#combined_x=combined_x[24:216, 24:216, 13:141]

combined_x=combined_x[56:184, 56:184, 13:141] #Crop to 128x128x128x4

#Do the same for mask
test_mask = test_mask[56:184, 56:184, 13:141]

print(combined_x.shape)
print(test_mask.shape)
```

```python
n_slice=random.randint(0, test_mask.shape[2])
n_slice = 64
print("slice nr.", n_slice)

plt.figure(figsize=(12, 8))

plt.subplot(151)
plt.imshow(combined_x[:,:,n_slice, 0], cmap='gray')
plt.title('Image t1')

plt.subplot(152)
plt.imshow(combined_x[:,:,n_slice, 1], cmap='gray')
plt.title('Image t2')

plt.subplot(153)
plt.imshow(combined_x[:,:,n_slice, 2], cmap='gray')
plt.title('Image t1ce')

plt.subplot(154)
plt.imshow(combined_x[:,:,n_slice, 3], cmap='gray')
plt.title('Image flair')

plt.subplot(155)
plt.imshow(test_mask[:,:,n_slice])
plt.title('Mask')
plt.show()
```

```python
# # # images lists harley
t1_list = sorted(glob.glob(os.path.join(TRAIN_DATASET_PATH, '*', '*' + '_t1.nii.gz')))
t2_list = sorted(glob.glob(os.path.join(TRAIN_DATASET_PATH, '*', '*' + '_t2.nii.gz')))
t1ce_list = sorted(glob.glob(os.path.join(TRAIN_DATASET_PATH, '*', '*' + '_t1ce.nii.gz')))
flair_list = sorted(glob.glob(os.path.join(TRAIN_DATASET_PATH, '*', '*' + '_flair.nii.gz')))

mask_list = sorted(glob.glob(os.path.join(TRAIN_DATASET_PATH, '*', '*' + '_seg.nii.gz')))
print(len(t1_list), len(t2_list), len(t1ce_list),len(flair_list), len(mask_list))
```

```python
#Each volume generates 18 64x64x64x4 sub-volumes. 
#Total 369 volumes = 6642 sub volumes
```

```python
t1_list[0].split('/')[-2]
```

```python
os.makedirs("BraTS2020_TrainingData/input_data_4channels/images", exist_ok=True) 
os.makedirs("BraTS2020_TrainingData/input_data_4channels/masks", exist_ok=True) 

os.makedirs("BraTS2020_TrainingData/input_data_4channels_excluded/images", exist_ok=True) 
os.makedirs("BraTS2020_TrainingData/input_data_4channels_excluded/masks", exist_ok=True) 

for img in range(len(t2_list)):   #Using t1_list as all lists are of same size
    print("Preparing image and masks number: ", img, t2_list[img].split('/')[-2])
      
    temp_image_t1=nib.load(t1_list[img]).get_fdata()
    temp_image_t1=scaler.fit_transform(temp_image_t1.reshape(-1, temp_image_t1.shape[-1])).reshape(temp_image_t1.shape)
    
    temp_image_t2=nib.load(t2_list[img]).get_fdata()
    temp_image_t2=scaler.fit_transform(temp_image_t2.reshape(-1, temp_image_t2.shape[-1])).reshape(temp_image_t2.shape)
   
    temp_image_t1ce=nib.load(t1ce_list[img]).get_fdata()
    temp_image_t1ce=scaler.fit_transform(temp_image_t1ce.reshape(-1, temp_image_t1ce.shape[-1])).reshape(temp_image_t1ce.shape)
   
    temp_image_flair=nib.load(flair_list[img]).get_fdata()
    temp_image_flair=scaler.fit_transform(temp_image_flair.reshape(-1, temp_image_flair.shape[-1])).reshape(temp_image_flair.shape)
        
    temp_mask=nib.load(mask_list[img]).get_fdata()
    temp_mask=temp_mask.astype(np.uint8)
    temp_mask[temp_mask==4] = 3  #Reassign mask values 4 to 3
    #print(np.unique(temp_mask))
    
    temp_combined_images = np.stack([temp_image_t1, temp_image_t2,temp_image_t1ce, temp_image_flair], axis=3)
    
    #Crop to a size to be divisible by 64 so we can later extract 64x64x64 patches. 
    #cropping x, y, and z
    temp_combined_images=temp_combined_images[56:184, 56:184, 13:141]
    temp_mask = temp_mask[56:184, 56:184, 13:141]
    
    val, counts = np.unique(temp_mask, return_counts=True)

    if (1 - (counts[0]/counts.sum())) > 0.01:  #At least 1% useful volume with labels that are not 0
        # print("Save image.")
        temp_mask= to_categorical(temp_mask, num_classes=4)
        np.save('BraTS2020_TrainingData/input_data_4channels/images/image_' + t2_list[img].split('/')[-2] + '.npy', temp_combined_images)
        np.save('BraTS2020_TrainingData/input_data_4channels/masks/mask_' + t2_list[img].split('/')[-2] + '.npy', temp_mask)

        # Calculating 1% of an image volume of size 128x128x128
        # image_size = np.prod(temp_mask.shape  # Total number of voxels in the image
        # one_percent = 0.01 * image_size  # 1% of the total number of voxels

    else:
        np.save('BraTS2020_TrainingData/input_data_4channels_excluded/images/image_' + t2_list[img].split('/')[-2] + '.npy', temp_combined_images)
        np.save('BraTS2020_TrainingData/input_data_4channels_excluded/masks/mask_' + t2_list[img].split('/')[-2] + '.npy', temp_mask)

        print("Exclude image.")   
```
